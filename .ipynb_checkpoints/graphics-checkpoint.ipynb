```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open('base_algo.txt', 'r') as f:
    input_data = f.read().split('\n')
```

```python
index = 0
ns = []
divisors = []
steps = []
times = []
while index < len(input_data):
    n, divisor, step = map(int, input_data[index].split())
    index += 1
    time = float(input_data[index])
    index += 1
    ns.append(n)
    divisors.append(divisor)
    steps.append(step)
    times.append(time)
```

```python
colors = [plt.cm.Spectral(i/float(np.max(steps))) for i in range(np.max(steps) + 1)]
```

```python
fig, ax = plt.subplots(figsize=(16, 9), dpi=80)
width = 0.5
N = 5
begin = 2**N - 2
end = 2**(N + 1) - 2

n_dict = {}
time_dict = {}
for n, step, time in zip(ns[begin:end], steps[begin:end], times[begin:end]):
    if step in n_dict:
        n_dict[step].append(n)
        time_dict[step].append(time)
    else :
        n_dict[step] = [n]
        time_dict[step] = [time]

for i in range(len(colors)):
    if i in n_dict:
        ax.bar(n_dict[i], time_dict[i], label=i, color=colors[i])
ax.set_ylabel('Секунд', fontsize=12)
ax.set_xlabel('N', fontsize=12)
plt.xlim(ns[begin:end][0] - 1, ns[begin:end][-1] + 1)
plt.title(f'Время выполнения алгоритма Шора для системы из {4 * (N + 1) + 3} кубитов', fontsize=16)
ax.set_xticks([i for i in range(ns[begin:end][0], ns[begin:end][-1] + 1)])
for i, val in enumerate(times[begin:end]):
    plt.text(i + begin + 2, val, round(val), horizontalalignment='center',
             verticalalignment='bottom', fontdict={'fontweight': 500, 'size': 8})
# plt.yscale('log')

plt.legend()
plt.savefig(f'./perf/{4 * (N + 1) + 3}.jpg')
plt.show()
```
