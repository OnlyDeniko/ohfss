```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open('base_algo.txt', 'r') as f:
    input_data = f.read().split('\n')
```

```python
index = 0
ns = []
angles = []
fs = []

while index < len(input_data):
    n, angle, f = map(float, input_data[index].split())
    n = int(n)
    index += 1
    ns.append(n)
    angles.append(abs(angle - 0.024))
    fs.append(f)
```

```python
start = 12
ns = ns[start:]
angles = angles[start:]
fs = fs[start:]
```

```python
start = np.min(ns)
end = np.max(ns)
vse = end - start + 1
colors = [plt.cm.Spectral(i/vse) for i in range(vse)]
```

```python
fig, ax = plt.subplots(figsize=(16, 9), dpi=80)

for i in range(len(ns)):
    ax.scatter(angles[i], fs[i], label=ns[i], color=colors[i])
ax.set_ylabel('Утечка', fontsize=12)
ax.set_xlabel('Модуль разности вычисленного угла и нужного угла', fontsize=12)
#plt.xlim(ns[begin:end][0] - 1, ns[begin:end][-1] + 1)
plt.title(f'Результат выполнения "базового" алгоритма', fontsize=16)
# ax.set_xticks([i for i in range(ns[begin:end][0], ns[begin:end][-1] + 1)])
# plt.yscale('log')

plt.legend()
# plt.savefig(f'./perf/{4 * (N + 1) + 3}.jpg')
plt.show()
```
